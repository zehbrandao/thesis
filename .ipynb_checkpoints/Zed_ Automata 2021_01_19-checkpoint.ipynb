## INNER WORKINGS

### General Purpose Functions

```python
def data_pruning(raw_data, columns_to_keep, columns_to_rename):
    raw_data = raw_data.reindex(columns=columns_to_keep)
    raw_data.rename(columns=columns_to_rename, inplace=True)
    
    return raw_data

def enforce_crs(geo_data, output_epsg):
    if geo_data.crs.to_epsg() != output_epsg:
        geo_data.to_crs(epsg=output_epsg, inplace=True)
    
    return geo_data


def remove_small_urban_clusters(geometry_of_tracts, municipalities):
    geometry_of_tracts = geometry_of_tracts.loc[geometry_of_tracts.zone=='URBANO']
    
    if not np.isin('Sao Paulo', municipalities):
        geometry_of_tracts = geometry_of_tracts.loc[geometry_of_tracts.district == geometry_of_tracts.]
        
    return geometry_of_tracts
```

### Getting Main Data

```python
def get_census_data(state, municipalities, census_year, csv_encoding='latin-1', separator=';'):
    
    path_to_census_data = pathlib.Path(r"C:\Users\brand\OneDrive\Documentos\Database\Census\{year} Universo\{state}\Basico_{state}.csv".format(year=census_year,
                                                                                                                                               state=state,))
    
    # Most variables below are dropped later, but I chose to keep them here because there's a specifica function to drop them,
    # so that allows me to change which ones to keep more easily 
    vars_best_as_strings = ['Cod_setor', 'Cod_Grandes RegiÃµes', 'Cod_UF', 'Cod_meso', 'Cod_micro', 
                            'Cod_RM', 'Cod_municipio', 'Cod_distrito', 'Cod_subdistrito', 'Cod_bairro'] 
    vars_best_as_strings = {var: str for var in vars_best_as_strings}
    
    census_data = pd.read_csv(path_to_census_data,
                              sep=separator,
                              encoding=csv_encoding,
                              dtype=vars_best_as_strings,)
    
    census_data.rename(columns={'Cod_setor': 'code_tract', 'Nome_do_municipio': 'municipality'}, inplace=True)
    
    census_data['municipality'] = census_data.municipality.map(lambda x: unidecode(x.title()))
    
    census_data = census_data.loc[census_data.municipality.isin(municipalities)]
    
    return census_data
    
    
    
def get_geometry_of_tracts(state, municipalities, census_year, tract_epsg=5641):
    
    geometry_of_tracts = read_census_tract(code_tract=state, year=census_year)
    geometry_of_tracts.to_crs(epsg=tract_epsg, inplace=True)
    
    geometry_of_tracts.rename(columns={'name_muni': 'municipality', 'name_district': 'district', 'name_subdistrict': 'subdistrict'}, inplace=True)
    geometry_of_tracts['code_muni'] = geometry_of_tracts.code_muni.map(lambda x: str(int(x)))
    geometry_of_tracts['municipality'] = geometry_of_tracts.municipality.map(lambda x: unidecode(x.title()))
    geometry_of_tracts['district'] = geometry_of_tracts.district.map(lambda x: unidecode(x.title()))
    
    geometry_of_tracts = geometry_of_tracts.loc[geometry_of_tracts.municipality.isin(municipalities)]
    
    geometry_of_tracts = remove_small_urban_clusters(geometry_of_tracts, municipalities)
    
    return geometry_of_tracts    

    
    
def georreference_census_data(state, municipalities, census_year, columns_to_keep, columns_to_rename, tract_epsg=5641):
    
    assert isinstance(municipalities, list), 'municipalities must be list-like'
    
    municipalities = [unidecode(municipality.title()) for municipality in municipalities]
    
    census_data = get_census_data(state, municipalities, census_year)
    geometry_of_tracts = get_geometry_of_tracts(state, municipalities, census_year, tract_epsg)
    
    tracts_with_data = pd.merge(geometry_of_tracts, census_data, how='inner', on='code_tract')
    
    tracts_with_data = data_pruning(tracts_with_data, columns_to_keep, columns_to_rename)
    
    return tracts_with_data
```

### Computing Densities

```python
def get_urban_footprint(tracts_with_data, urban_footprint_year):
    
    urban_footprint = read_urban_area(year=urban_footprint_year) # Only available for 2005 and 2015
    urban_footprint['code_muni'] = urban_footprint.code_muni.map(lambda x: str(int(x)))
    
    muni_codes = tracts_with_data.code_muni.unique()
    
    urban_footprint = urban_footprint.loc[urban_footprint.code_muni.isin(muni_codes)]
    
    return urban_footprint



def compute_urban_area_within_tracts(tracts_with_data, urban_footprint_year):
    
    urban_footprint = get_urban_footprint(tracts_with_data, urban_footprint_year)
    
    tracts_epsg = tracts_with_data.crs.to_epsg()
    urban_footprint.to_crs(epsg=tracts_epsg, inplace=True)
    
    tracts_with_data = gpd.clip(tracts_with_data, urban_footprint)
    tracts_with_data['tract_urban_area_km2'] = tracts_with_data.geometry.map(lambda x: x.area * 1e-6)
    
    return tracts_with_data
    
    

def compute_tract_urban_density(tracts_with_data, vars_to_get_density_of, urban_footprint_year):
    
    assert isinstance(vars_to_get_density_of,list), "vars_to_use must be list-like"
    
    tracts_with_data = compute_urban_area_within_tracts(tracts_with_data, urban_footprint_year)
    
    for var in vars_to_get_density_of:
        tracts_with_data[f'{var}_density_km2'] = tracts_with_data[var] / tracts_with_data.tract_urban_area_km2
    
    return tracts_with_data
```

### Defining Study Area Using H3 Hexagons

```python
def get_study_area_boundaries(study_area_tracts, census_year):
    """
    To be used in get_H3_labels_within_study_area
    """
    epsg = study_area_tracts.crs.to_epsg()
    
    municipalities_in_study_area = study_area_tracts.code_muni.astype(int).unique()
    study_area_boundaries = read_municipality(year=census_year)
    study_area_boundaries = study_area_boundaries.loc[study_area_boundaries.code_muni.isin(municipalities_in_study_area)]
    study_area_boundaries.to_crs(epsg=epsg, inplace=True)
    
    study_area_envelope = study_area_tracts.geometry.unary_union.envelope
    study_area_envelope = gpd.GeoSeries(study_area_envelope, name='geometry', crs=f'EPSG:{epsg}')
    study_area_envelope = gpd.GeoDataFrame(study_area_envelope, geometry='geometry')
    
    study_area_boundaries = gpd.overlay(study_area_boundaries, study_area_envelope, how='intersection')
    
    return study_area_boundaries
    


def get_H3_labels_within_study_area(hexagon_size, study_area_tracts, census_year):
    """
    To be used within get_hexagon_geometries
    """
    
    study_area_boundaries = get_study_area_boundaries(study_area_tracts, census_year)
    study_area_boundaries.to_crs(epsg=4326, inplace=True) # For compatibility reasons -> EPSG:4326 seems the CRS used by h3 library
    study_area_boundaries = study_area_boundaries.geometry
    study_area_boundaries = study_area_boundaries.explode()
    
    h3_labels = []
    for boundary in study_area_boundaries:
        temp = mapping(boundary)
        temp['coordinates'] = [
            
            [[lon,lat] for lon,lat in shape] for shape in temp['coordinates'] # Ugly, but h3 demands it so
        ]
        
        h3_labels.extend(
            h3.polyfill(temp, hexagon_size, geo_json_conformant=True) # Memento: GeoJSON is in lon,lat format
        )
        
    return h3_labels
    
    
    
def get_hexagon_geometries(hexagon_size, study_area_tracts, urban_footprint_year, output_epsg=5641):
    
    h3_labels = get_H3_labels_within_study_area(hexagon_size, study_area_tracts, urban_footprint_year)
    hex_col_name = f'hex_{hexagon_size}'
    hexagons = pd.DataFrame(h3_labels, columns=[hex_col_name])
    
    no_of_labels = len(h3_labels)
    hexagons.drop_duplicates(inplace=True)
    print('Sanity check:\n', f"{no_of_labels} H3 labels, with resolution {hexagon_size}, resulted in dataframe with {len(hexagons)} rows, after dropping duplicates")
    
    hexagons['hex_vertices_coords'] = hexagons[hex_col_name].map(lambda x: h3.h3_to_geo_boundary(x, geo_json=True)) # Memento (again): GeoJSON is in lon,lat format
    hexagons['geometry'] = hexagons.hex_vertices_coords.map(lambda x: Polygon(x))
    hexagons.drop(columns='hex_vertices_coords', inplace=True)
    
    hexagons = gpd.GeoDataFrame(hexagons, crs='EPSG:4326', geometry='geometry') # Memento: EPSG:4326 refers to WSG84 crs, because of H3
    hexagons.to_crs(epsg=output_epsg, inplace=True)
    hexagons['hexagon_area_km2'] = hexagons.geometry.map(lambda x: x.area * 1e-6)
    
    return hexagons
```

### Transforming Data: From Tracts to H3 Hexagons

```python
def summarize_geodata_with_groupby(group):
    """
    To be used within transform_data_between_geometries
    This might generalise further, but it assumes that each in vars_to_transform is density-like, i.e., thing/area
    """
   
    areas_km2 = group.geometry.map(lambda x: x.area * 1e-6)
    vars_to_transform = list(group.columns)
    vars_to_transform.remove('geometry')
    
    temp = {}
    for var in vars_to_transform:
        temp[var] = (areas_km2 * group[var]).sum()
        
    series = pd.Series(temp, index=vars_to_transform)
    
    return series



def transform_data_between_geometries(source, source_labels, destination, destination_labels, destination_geom_areas, vars_to_transform, output_epsg=5641):
    
    source = enforce_crs(source, output_epsg)
    destination = enforce_crs(destination, output_epsg)
    
    intersection = gpd.overlay(source,
                               destination,
                               how='intersection',
                              )
    
    vars_to_transform.append('geometry')
    intersection = intersection.groupby(destination_labels)[vars_to_transform].apply(summarize_geodata_with_groupby)
    vars_to_transform.remove('geometry')
    
    destination = destination.merge(intersection,
                                    how='left',
                                    left_on=destination_labels,
                                    right_index=True,
                                   )
    
    replacements = {var: 0 for var in vars_to_transform}
    destination.fillna(replacements, inplace=True)
    
    for var in vars_to_transform:
        destination[var] = destination[var] / destination[destination_geom_areas]
    
    return destination



def transport_data_between_geometries(source, destination, destination_labels, vars_to_transport, output_epsg=5641):
    
    geometry = destination.set_index(destination_labels).geometry
    centres = geometry.map(lambda x: x.centroid)
    centres = gpd.GeoDataFrame(centres)
    
    centres.set_crs(epsg=output_epsg, inplace=True)
    source = enforce_crs(source, output_epsg)
    
    hexagons_by_district = gpd.sjoin(centres,
                                     source,
                                     how='left',
                                     op='within',
                              )
    
    hexagons_by_district = hexagons_by_district.loc[:,list(vars_to_transport)]
    destination = destination.merge(hexagons_by_district, how='left', left_on=destination_labels, right_index=True)
    
    return destination
```

### The Fractal Forms of Urban Land

We here try three different version of the standard radial dimension, all concerning populational density:

- First, as in the classical definition presented, for example, in Box 5.1 of White et al.'s Modeling Cities and Regions as Complex Systems.
    - Also refer, e.g., to https://doi.org/10.1007/s10109-006-0026-9
    
- Then, we move to the previous measure's generalisation, i.e., the *correlation dimension*

- Finally, an attemp is made to implement the radial dimension as per Thomas, Frankhauser and Keersmaecker (2007)
    - Basically it *is* the radial dimension, but applied to different spatial parcels
    - reference: https://hal.archives-ouvertes.fr/hal-00731651

#### The Radial Dimension

```python
def get_study_area_smallest_extension(hexagons):
    # TO DO: this operation is already performed elsewhere (get_study_area_boundaries). Try to think of way to only do it once
    
    x_min, y_min, x_max, y_max = hexagons.geometry.unary_union.bounds
    
    # This assumes a projected coordinate system (specifically, assumes distances in metres)
    x_distance = x_max - x_min
    y_distance = y_max - y_min
    
    extension = min(x_distance, y_distance)
    
    return extension



def get_hexagon_height(hexagon_size, hexagons):
    # This assumes a projected coordinate system (specifically, assumes distances in metres)
    sample = hexagons.geometry[0]
    perimeter = sample.length
    edge = perimeter/6
    height = (edge/2)*(3/np.sqrt(3))
    
    return height



def get_city_centre_coordinates(state, municipality, year=2010):
    #TO DO: Generalize input year in a way that given a year, the function chooses the closest year among those available
    # Municipal seat (location of city hall) assumed to be at a city's original centre (it usually is)
    municipal_seats = read_municipal_seat(year=year).to_crs(epsg=4326) # Memento: to correspond to H3's base crs
    municipal_seats['name_muni'] = municipal_seats.name_muni.map(lambda x: unidecode(x.title()))
    
    mask = municipal_seats.name_muni.isin(municipality) & municipal_seats.abbrev_state.isin([state])
    municipal_seat_coords = municipal_seats.loc[mask]
    municipal_seat_coords = municipal_seat_coords.geometry.iloc[0]
    
    return municipal_seat_coords.x, municipal_seat_coords.y
    

    
def get_central_hexagon_label(state, municipality, hexagon_size):
    lon,lat = get_city_centre_coordinates(state, municipality, year=2010)
    hexagon_label = h3.geo_to_h3(lat, lon, hexagon_size)
    
    return hexagon_label



def get_rings_from_centre(state, municipality, hexagon_size, number_of_rings):    
    """
    Output: list of sets. Each set is a k-order group of neighbors 
    """
    hexagon_label = get_central_hexagon_label(state, municipality, hexagon_size)
    rings_around_center = h3.hex_range_distances(hexagon_label, number_of_rings)
    
    return rings_around_center


def get_neighborhood(state, municipality, hexagon_size, number_of_rings):
    """
    Basically, function takes an n number o rings centered at a given hexagon and flattens them into a list.
    Output: list of hexagons defining of a contiguous neighborhood of hexagons.
    """
    rings = get_rings_from_centre(state, municipality, hexagon_size, number_of_rings)
    
    neighborhood = set()
    for ring in rings:
        neighborhood = neighborhood.union(ring)
        
    return list(neighborhood)



def compute_mean_neighborhood_density(neighborhood, hexagon_area_column, dimension):
    mean_neighborhood_density = (neighborhood[dimension] * neighborhood[hexagon_area_column]).sum()/ neighborhood[hexagon_area_column].sum()
    
    return mean_neighborhood_density


def compute_radial_dimension(state, municipality, hexagons, hexagon_size, hexagon_area_column, dimension):
    extension = get_study_area_smallest_extension(hexagons)
    hexagon_height = get_hexagon_height(hexagon_size, hexagons)
    
    no_hexagons = len(hexagons)
    
    k = 1
    counter = 0
    distance_from_centre = k * hexagon_height
    radial_dimensions = {}
    while distance_from_centre < extension/2:
        k += 1
        neighborhood = get_neighborhood(state, municipality, hexagon_size, k)
        neighborhood = hexagons.loc[hexagons[f'hex_{hexagon_size}'].isin(neighborhood)]
        neighborhood = neighborhood.loc[:,[dimension, hexagon_area_column, 'geometry']] # TO DO: consider not to hardcode this
        radial_dimensions[k] = {'dist': distance_from_centre,
                                'measure': compute_mean_neighborhood_density(neighborhood, hexagon_area_column, dimension)}
    
        distance_from_centre = k * hexagon_height
        
        counter += 1
        if counter > no_hexagons/2:
            print(f"To many loops ({counter}): Something's probably wrong")
            break
    
    radial_dimensions = pd.DataFrame.from_dict(radial_dimensions, orient='index')
    
    return radial_dimensions
```

```python
foo = radial_dimensions.copy()
waldo = foo.copy()
waldo = waldo.loc[foo.dist<10000]
waldo['dist'] = waldo.dist.map(np.log)
waldo['measure'] = waldo.measure.map(np.log)
waldo.plot(x='dist', y='measure')
```

```python
from scipy.optimize import curve_fit
import scipy.odr
import scipy.stats


def func(dist, k, D):
    return np.log(k) + D*np.log(dist)

x = waldo['dist']
y = waldo['measure']
parameters, cov = curve_fit(func, x, y)


def f_wrapper_for_odr(beta, dist): # parameter order for odr
    return func(dist, *beta)

model = scipy.odr.odrpack.Model(f_wrapper_for_odr)
data = scipy.odr.odrpack.Data(x,y)
myodr = scipy.odr.odrpack.ODR(data, model, beta0=parameters,  maxit=0)
myodr.set_job(fit_type=2)
parameterStatistics = myodr.run()
df_e = len(x) - len(parameters) # degrees of freedom, error
cov_beta = parameterStatistics.cov_beta # parameter covariance matrix from ODR
sd_beta = parameterStatistics.sd_beta * parameterStatistics.sd_beta
ci = []
t_df = scipy.stats.t.ppf(0.975, df_e)
ci = []
for i in range(len(parameters)):
    ci.append([parameters[i] - t_df * parameterStatistics.sd_beta[i], parameters[i] + t_df * parameterStatistics.sd_beta[i]])

tstat_beta = parameters / parameterStatistics.sd_beta # coeff t-statistics
pstat_beta = (1.0 - scipy.stats.t.cdf(np.abs(tstat_beta), df_e)) * 2.0    # coef. p-values
pstat_beta
```
